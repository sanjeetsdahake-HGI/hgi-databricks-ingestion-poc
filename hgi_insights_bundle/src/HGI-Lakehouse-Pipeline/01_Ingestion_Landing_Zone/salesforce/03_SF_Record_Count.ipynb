```python
sf_files_count = 0
keys = ["account", "campaign", "campaignmember", "contact", "opportunity", "task", "user"]  

for key in keys:
    retrieved_count = dbutils.jobs.taskValues.get(
        taskKey=f"hgi_crm_incremental_load_{key}", 
        key=key, 
        default=0
    )
    sf_files_count += retrieved_count
    print(f"Key: {key}, Retrieved Count: {retrieved_count}, Greater than 0: {retrieved_count > 0}, sf_files_count: {sf_files_count}")

dbutils.jobs.taskValues.set(key="sf_files_count",  value=str(sf_files_count))
```
