"To demonstrate how Delta Lake handles transactions, I've created a quick clone of our Accounts table. Currently, this table is at Version 0, and we can see our baseline total record count."

```python
%sql
CREATE TABLE hgi.silver.accounts_demo 
AS SELECT * FROM hgi.silver.accounts;
```

```python
%sql
-- 2. Let's check the current version of our new demo table.
-- It should start at Version 0.
DESCRIBE HISTORY hgi.silver.accounts_demo;
```

```python
%sql
-- 3. Let's get a baseline count of our records
SELECT count(*) AS total_records FROM hgi.silver.accounts_demo;
```

```python
spark.sql("DELETE FROM hgi.silver.accounts_demo")
```

```python
%sql
SELECT count(*) AS total_records_after_delete 
FROM hgi.silver.accounts_demo;
```

```python
%sql
DESCRIBE HISTORY hgi.silver.accounts_demo;
```

```python
spark.sql("RESTORE TABLE hgi.silver.accounts_demo TO VERSION AS OF 0")
count = spark.table("hgi.silver.accounts_demo").count()
print(f"Number of records after restore: {count}")
```
