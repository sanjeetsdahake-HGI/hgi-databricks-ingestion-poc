#### Key Auto Loader Features:

- **Schema Inference and Evolution:**  
  Auto Loader can automatically detect the schema of incoming data, reducing the need for manual definition. It can also detect new columns and evolve the table schema, allowing for the ingestion of evolving data formats without breaking pipelines.

- **Rescued Data Column:**  
  To prevent data loss, Auto Loader can capture malformed or mismatched records into a dedicated "rescued data" column. This ensures that all data is ingested, allowing for later analysis and debugging of data quality issues.

- **Efficient File Detection Modes:**  
  - *Directory Listing:* Default mode; incrementally scans directories for new files using APIs.  
  - *File Notification:* Uses cloud-native notification services (like SQS, SNS, Azure Event Grid) for faster detection in high-volume scenarios, reducing costs associated with listing large directories.

- **Scalability and Performance:**  
  Designed to handle billions of files and process millions of files per hour, making it suitable for large-scale data lake ingestion.

- **Exactly-Once Processing:**  
  Uses checkpointing and Spark Structured Streaming to guarantee that data is processed exactly once, ensuring data consistency.

- **Supported File Formats:**  
  Ingests CSV, JSON, XML, Parquet, Avro, ORC, Text, and Binary File formats.

- **Integration with Delta Live Tables (DLT):**  
  Easily integrates into DLT pipelines for declarative, high-quality data ingestion.
