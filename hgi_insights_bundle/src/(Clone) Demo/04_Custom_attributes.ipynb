```python
from pyspark.sql.functions import col, row_number
from pyspark.sql.window import Window

window_spec = Window.partitionBy("tenant").orderBy(col("id"))

df = (
    spark.table("hgi.silver.accounts")
    .filter((col("tenant").isin("1", "2")) & (col("custom_metadata").isNotNull()))
    .select("tenant", "id", "custom_metadata")
    .withColumn("rn", row_number().over(window_spec))
    .filter(col("rn") <= 5)
    .drop("rn")
)
display(df)
```
