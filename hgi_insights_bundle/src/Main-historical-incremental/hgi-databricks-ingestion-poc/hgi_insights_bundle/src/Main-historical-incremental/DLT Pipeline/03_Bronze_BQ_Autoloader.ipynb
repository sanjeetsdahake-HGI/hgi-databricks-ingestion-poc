```python
df = spark.readStream \
  .format("cloudFiles") \
  .option("cloudFiles.format", "parquet") \
  .option("cloudFiles.schemaLocation", "s3://hgi-databricks-data-lakehouse-dev/checkpoints/bq_events_schema") \
  .load("s3://hgi-databricks-data-lakehouse-dev/landing-zone/bigquery/cust_001/events")

(df.writeStream
  .format("delta")
  .option("checkpointLocation","s3://hgi-databricks-data-lakehouse-dev/checkpoints/bq_events")
  .trigger(availableNow=True)
  .start("s3://hgi-databricks-data-lakehouse-dev/layers/bronze/tables/bigquery/events"))
```
